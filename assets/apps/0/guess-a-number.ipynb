#### Basic Coding Ideas:
* You can assign a value to a variable in the following way:
```
a = 3
```
This will give the variable `a` the value `3`, which can then be used in future calculations such as:
```
print(a + 3)
```
which would return `6`.


* Unlike many other programming languages such as Visual Basic, there is no need to declare variables to have a certain *type* in Python. Python will automatically figure out whether you are specifying an integer, a real number or a string (if it's in quotes). You can check the *type* of a variable with
```
type(a)
```
Variables can be named any combination of letters and numbers both lower case and upper case but must start with a letter. Python is case sensitive, so `A` is not the same as `a`!


* Comments in Python are achieved with a `#` character. This means that anything written after the `#` in a line will be ignored. You should always write comments explaining what complicated lines do and write code in a coherent way. This is not exclusive to your code being readable by someone else either - if you write a code that is not properly commented and don’t look at it for a few months, you will not have any idea what it does the next time you look at it! This has happened to me so many times that I am determined to not let you make that mistake.

```
print('Hello World') #This is a comment
```


* **Indentation** in Python is **crucial**. These two code pieces are not the same, and in fact the second one will give a syntax error:

```
for j in range(1,10):
    print(j)


for j in range(1,10):
print(j)
```

* If there is nothing else you take from this document, let the thing you learn be how to use **for loops** and **if statements**. These are central to most programs you will write during this course and usually play an important role in many other programs. The very simple example below illustrates what these statements do:

```
for j in range(1,10):
    print(j) #display j at each point in the loop

for j in range(1,10):
    if j == 5 or j == 7
        print(j) # will only print 5 and 7

```

````python
import random

#generate random number and store as variable rnd
rnd = random.randint(1,10)

#take user input and store in variable uans
print('I am thinking of a whole number between 1 and 10. I want you to try and guess it!')
uans = input('Guess a whole number: ')

#repeat until number is guessed correctly
while (rnd != uans):
    if (rnd > uans):
        #tell the player they need to guess a higher number, because theirs was too low
        uans = input('Higher! Guess again: ')
    elif (rnd < uans):
        #tell the player they need to guess a lower number, because theirs was too high
        uans = input('Lower! Guess again: ')
        
print('Congratulations! You guessed the number correctly!')
````
